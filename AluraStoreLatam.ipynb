### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

```python
ingreso_total_1= tienda['Precio'].sum()
ingreso_total_2= tienda2['Precio'].sum()
ingreso_total_3= tienda3['Precio'].sum()
ingreso_total_4= tienda4['Precio'].sum()


print(f"El ingreso total de la tienda   es: {ingreso_total_1:.2f}")
print(f"El ingreso total de la tienda 2 es: {ingreso_total_2:.2f}")
print(f"El ingreso total de la tienda 3 es: {ingreso_total_3:.2f}")
print(f"El ingreso total de la tienda 4 es: {ingreso_total_4:.2f}")
```

```python
import matplotlib.pyplot as plt
```

```python
## SE MOSTRARA UN INGRESO TOTAL DE LA SUMATORIA DE TODAS LAS TIENDAS Y AHI SE MOSTRARA QUE PORCENTAJE DE ESE TOTAL REPRESENTA CADA TIENDA

nombres_tiendas = ['Tienda ', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ingresos = [ingreso_total_1, ingreso_total_2, ingreso_total_3, ingreso_total_4]

plt.figure(figsize=(4, 4))
plt.pie(ingresos, labels=nombres_tiendas, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral', 'lightgreen', 'gold'])
plt.title('Porcentaje del ingreso TOTAL por tienda')
plt.axis('equal') # Para asegurar que el círculo sea un círculo
plt.show()
```

```python
import matplotlib.ticker as mticker

df_ingresos_comparacion = pd.DataFrame({'Ingresos': ingresos}, index=nombres_tiendas)

ax = df_ingresos_comparacion.plot(kind='bar', figsize=(6, 6), ylabel='Ingresos totales', title='Comparación de Ingresos Totales por Tienda', legend=False)

# Formatear el eje Y para mostrar los valores en millones
formatter = mticker.FuncFormatter(lambda x, p: f'{x/1e6:.1f}M')
ax.yaxis.set_major_formatter(formatter)

plt.tight_layout()
plt.show()
```

# 2. Ventas por categoría

```python
venta_categoria_1 = tienda['Categoría del Producto'].value_counts()
venta_categoria_2 = tienda2['Categoría del Producto'].value_counts()
venta_categoria_3 = tienda3['Categoría del Producto'].value_counts()
venta_categoria_4 = tienda4['Categoría del Producto'].value_counts()
```

```python
display(venta_categoria_1)
display(venta_categoria_2)
display(venta_categoria_3)
display(venta_categoria_4)
```

```python
df_ventas_comparacion = pd.DataFrame({'Tienda 1': venta_categoria_1, 'Tienda 2': venta_categoria_2,'Tienda 3': venta_categoria_3, 'Tienda 4': venta_categoria_4}).fillna(0)

ax = df_ventas_comparacion.plot(kind='barh', figsize=(10, 6), xlabel='Cantidad', ylabel='Categoría del Producto', title='Ventas por Categoría', legend=True)
plt.tight_layout()
plt.show()
```

# 4. Productos más y menos vendidos

```python
cant_productos = tienda['Producto'].value_counts()
producto_mayor_ventas = cant_productos.idxmax()
ventas_maximas = cant_productos.max()

cant_productos2 = tienda2['Producto'].value_counts()
producto_mayor_ventas2 = cant_productos2.idxmax()
ventas_maximas2 = cant_productos2.max()

cant_productos3 = tienda3['Producto'].value_counts()
producto_mayor_ventas3 = cant_productos3.idxmax()
ventas_maximas3 = cant_productos3.max()

cant_productos4 = tienda4['Producto'].value_counts()
producto_mayor_ventas4 = cant_productos4.idxmax()
ventas_maximas4 = cant_productos4.max()

print(f"El producto con mayores ventas de la tienda   es: '{producto_mayor_ventas}' con {ventas_maximas} unidades vendidas.")
print(f"El producto con mayores ventas de la tienda 2 es: '{producto_mayor_ventas2}' con {ventas_maximas2} unidades vendidas.")
print(f"El producto con mayores ventas de la tienda 3 es: '{producto_mayor_ventas3}' con {ventas_maximas3} unidades vendidas.")
print(f"El producto con mayores ventas de la tienda 4 es: '{producto_mayor_ventas4}' con {ventas_maximas4} unidades vendidas.")
```

```python
producto_menor_ventas = cant_productos.idxmin()
ventas_minimas = cant_productos.min()

producto_menor_ventas2 = cant_productos2.idxmin()
ventas_minimas2 = cant_productos2.min()

producto_menor_ventas3 = cant_productos3.idxmin()
ventas_minimas3 = cant_productos3.min()

producto_menor_ventas4 = cant_productos4.idxmin()
ventas_minimas4 = cant_productos4.min()

print(f"El producto con menores ventas de la tienda   es: '{producto_menor_ventas}' con {ventas_minimas} unidades vendidas.")
print(f"El producto con menores ventas de la tienda 2 es: '{producto_menor_ventas2}' con {ventas_minimas2} unidades vendidas.")
print(f"El producto con menores ventas de la tienda 3 es: '{producto_menor_ventas3}' con {ventas_minimas3} unidades vendidas.")
print(f"El producto con menores ventas de la tienda 4 es: '{producto_menor_ventas4}' con {ventas_minimas4} unidades vendidas.")
```
